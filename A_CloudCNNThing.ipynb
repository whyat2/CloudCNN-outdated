```python
FIRSTLAYERS = 2
SECONDLAYERS = 2

FILTERSSTART = 28

FILTERSIZE = (10, 10)
FILTERTWOSIZE = (9, 9)

FILTERSTRIDES = (2, 2)
FILTERTWOSTRIDES = (5, 5)
```

```python
def generate_CNN():
    #model = tf.keras.Sequential([
    #    keras.layers.Input(shape=(150, 150, 3)),
    #    keras.layers.Conv2D(64, (10, 10), strides=(2, 2), padding='same'),
    #    keras.layers.LeakyReLU(),
    #
    #    keras.layers.Conv2D(64, (7, 7), strides=(2, 2), padding='same'),
    #    keras.layers.LeakyReLU(),
    #
    #    keras.layers.Conv2D(128, (10, 10), strides=(5, 5), padding='same'),
    #    keras.layers.LeakyReLU(),
    #    keras.layers.Flatten(),
    #    keras.layers.Dense(6, activation='sigmoid'),
    #    keras.layers.Dense(6, activation='softmax')
    #])
    model = tf.keras.Sequential()
    model.add(keras.layers.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, IMAGE_LAYERS)))
    for i in range(0, FIRSTLAYERS):
        model.add(keras.layers.Conv2D(int(FILTERSSTART + i * FILTERSSTART / 2), FILTERSIZE, FILTERSTRIDES, padding='same'))
        model.add(keras.layers.LeakyReLU())
    
    for i in range(0, SECONDLAYERS):
        model.add(keras.layers.Conv2D(FILTERSSTART + i * FILTERSSTART + FILTERSSTART * FIRSTLAYERS, FILTERSTRIDES, padding='same'))
        model.add(layers.LeakyReLU())
    
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(IMAGE_CATEGORIES, activation='softmax'))
    return model
```
